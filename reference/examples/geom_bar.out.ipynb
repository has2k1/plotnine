# Bar chart

```python

from plotnine import ggplot, aes, geom_bar, coord_flip, theme_classic
from plotnine.data import mpg
```

### Bar chart

*geom_bar() makes the height of the bar proportional to the number of
cases in each group*

```python
mpg.head()
```

```python
ggplot(mpg) + geom_bar(aes(x="class"))
```

```python
(
    ggplot(mpg)
    + geom_bar(aes(x="class", fill="drv"))
)
```

```python
(
    ggplot(mpg) 
    + geom_bar(aes(x="class", fill="drv"))
    + coord_flip()
    + theme_classic()
)
```
